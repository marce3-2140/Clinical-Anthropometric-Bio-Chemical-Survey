```python
import pandas as pd

# Aseg√∫rate de que 'df' es tu DataFrame
# Crear una columna para el a√±o de contrataci√≥n
df['A√±o de Contrataci√≥n'] = pd.to_datetime(df['hire_date']).dt.year

# Calcular el n√∫mero total de empleados por a√±o y g√©nero
contrataciones_por_a√±o_genero = df.groupby(['A√±o de Contrataci√≥n', 'gender']).size().unstack(fill_value=0)

# Cambiar nombres de columnas a 'Female' y 'Male' si es necesario
#contrataciones_por_a√±o_genero.columns = ['Female', 'Male']
```

```python
import matplotlib.pyplot as plt

# Crear la gr√°fica de l√≠neas
plt.figure(figsize=(10, 6))
plt.plot(contrataciones_por_a√±o_genero.index, contrataciones_por_a√±o_genero['F'], marker='o', label='Mujeres')
plt.plot(contrataciones_por_a√±o_genero.index, contrataciones_por_a√±o_genero['M'], marker='o', label='Hombres')

# A√±adir t√≠tulo y etiquetas
plt.title('Contrataciones por A√±o y G√©nero')
plt.xlabel('A√±o de Contrataci√≥n')
plt.ylabel('N√∫mero de Contrataciones')
plt.legend()

# Mostrar la gr√°fica
plt.grid(True)
plt.show()
```

```python
df['hire_date'] = pd.to_datetime(df['hire_date'])

# Encontrar el √∫ltimo a√±o de contrataci√≥n
ultimo_a√±o_contratacion = df['hire_date'].dt.year.max()

print(f"El √∫ltimo a√±o de contrataci√≥n registrado es: {ultimo_a√±o_contratacion}")
```

```python
# Extraer el d√≠a del mes como una nueva columna
df['Day_of_Month'] = df['hire_date'].dt.day

# Calcular el n√∫mero de contrataciones por d√≠a del mes y g√©nero
contrataciones_por_dia_genero = df.groupby(['Day_of_Month', 'gender']).size().unstack(fill_value=0)
```

```python
import matplotlib.pyplot as plt

# Ordenar los datos por d√≠a del mes
contrataciones_por_dia_genero = contrataciones_por_dia_genero.sort_index()

# Configurar la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar los datos
for gender in contrataciones_por_dia_genero.columns:
    ax.plot(contrataciones_por_dia_genero.index, contrataciones_por_dia_genero[gender], marker='o', label=gender)

# A√±adir t√≠tulo y etiquetas
ax.set_title('N√∫mero de Contrataciones por D√≠a del Mes y G√©nero')
ax.set_xlabel('D√≠a del Mes')
ax.set_ylabel('N√∫mero de Contrataciones')
ax.legend()

# Mostrar la gr√°fica
plt.grid(True)
plt.show()
```

```python
# Extraer el d√≠a del mes como una nueva columna
df['Day_of_Month'] = df['hire_date'].dt.day

# Calcular el n√∫mero de contrataciones por d√≠a del mes y g√©nero
contrataciones_por_dia_genero = df.groupby(['Day_of_Month', 'gender']).size().unstack(fill_value=0)

# Encontrar el d√≠a del mes con el mayor n√∫mero de contrataciones para ambos g√©neros
max_contrataciones_dia = contrataciones_por_dia_genero.sum(axis=1).idxmax()

print(f"El d√≠a del mes con el mayor n√∫mero de contrataciones para ambos g√©neros es: {max_contrataciones_dia}")
```

```python
# Extraer el d√≠a del mes como una nueva columna
df['Day_of_Month'] = df['hire_date'].dt.day

# Calcular el n√∫mero de contrataciones por d√≠a del mes y g√©nero
contrataciones_por_dia_genero = df.groupby(['Day_of_Month', 'gender']).size().unstack(fill_value=0)

# Encontrar el d√≠a del mes con el mayor n√∫mero de contrataciones para ambos g√©neros
max_contrataciones_total = contrataciones_por_dia_genero.sum(axis=1).max()

# Filtrar los d√≠as del mes que tienen la mayor cantidad de contrataciones para ambos g√©neros
dias_max_contrataciones = contrataciones_por_dia_genero.index[contrataciones_por_dia_genero.sum(axis=1) == max_contrataciones_total].tolist()

print(f"Los d√≠as del mes con la mayor cantidad de contrataciones para ambos g√©neros ({max_contrataciones_total} contrataciones) son: {dias_max_contrataciones}")
```
