```python
import pandas as pd

# Asegúrate de que 'df' es tu DataFrame
# Crear una columna para el año de contratación
df['Año de Contratación'] = pd.to_datetime(df['hire_date']).dt.year

# Calcular el número total de empleados por año y género
contrataciones_por_año_genero = df.groupby(['Año de Contratación', 'gender']).size().unstack(fill_value=0)

# Cambiar nombres de columnas a 'Female' y 'Male' si es necesario
#contrataciones_por_año_genero.columns = ['Female', 'Male']
```

```python
import matplotlib.pyplot as plt

# Crear la gráfica de líneas
plt.figure(figsize=(10, 6))
plt.plot(contrataciones_por_año_genero.index, contrataciones_por_año_genero['F'], marker='o', label='Mujeres')
plt.plot(contrataciones_por_año_genero.index, contrataciones_por_año_genero['M'], marker='o', label='Hombres')

# Añadir título y etiquetas
plt.title('Contrataciones por Año y Género')
plt.xlabel('Año de Contratación')
plt.ylabel('Número de Contrataciones')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
df['hire_date'] = pd.to_datetime(df['hire_date'])

# Encontrar el último año de contratación
ultimo_año_contratacion = df['hire_date'].dt.year.max()

print(f"El último año de contratación registrado es: {ultimo_año_contratacion}")
```

```python
# Extraer el día del mes como una nueva columna
df['Day_of_Month'] = df['hire_date'].dt.day

# Calcular el número de contrataciones por día del mes y género
contrataciones_por_dia_genero = df.groupby(['Day_of_Month', 'gender']).size().unstack(fill_value=0)
```

```python
import matplotlib.pyplot as plt

# Ordenar los datos por día del mes
contrataciones_por_dia_genero = contrataciones_por_dia_genero.sort_index()

# Configurar la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar los datos
for gender in contrataciones_por_dia_genero.columns:
    ax.plot(contrataciones_por_dia_genero.index, contrataciones_por_dia_genero[gender], marker='o', label=gender)

# Añadir título y etiquetas
ax.set_title('Número de Contrataciones por Día del Mes y Género')
ax.set_xlabel('Día del Mes')
ax.set_ylabel('Número de Contrataciones')
ax.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
# Extraer el día del mes como una nueva columna
df['Day_of_Month'] = df['hire_date'].dt.day

# Calcular el número de contrataciones por día del mes y género
contrataciones_por_dia_genero = df.groupby(['Day_of_Month', 'gender']).size().unstack(fill_value=0)

# Encontrar el día del mes con el mayor número de contrataciones para ambos géneros
max_contrataciones_dia = contrataciones_por_dia_genero.sum(axis=1).idxmax()

print(f"El día del mes con el mayor número de contrataciones para ambos géneros es: {max_contrataciones_dia}")
```

```python
# Extraer el día del mes como una nueva columna
df['Day_of_Month'] = df['hire_date'].dt.day

# Calcular el número de contrataciones por día del mes y género
contrataciones_por_dia_genero = df.groupby(['Day_of_Month', 'gender']).size().unstack(fill_value=0)

# Encontrar el día del mes con el mayor número de contrataciones para ambos géneros
max_contrataciones_total = contrataciones_por_dia_genero.sum(axis=1).max()

# Filtrar los días del mes que tienen la mayor cantidad de contrataciones para ambos géneros
dias_max_contrataciones = contrataciones_por_dia_genero.index[contrataciones_por_dia_genero.sum(axis=1) == max_contrataciones_total].tolist()

print(f"Los días del mes con la mayor cantidad de contrataciones para ambos géneros ({max_contrataciones_total} contrataciones) son: {dias_max_contrataciones}")
```
